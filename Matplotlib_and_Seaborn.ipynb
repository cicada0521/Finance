```python
# 1. Create the following DataFrame Sales containing year wise sales figures for five sales persons in INR.
# Use the years as column labels, and sales person names as row labels.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create sales data
sales_data = {
    '2014': [100.5, 150.8, 200.9, 30000, 40000],
    '2015': [12000, 18000, 22000, 30000, 45000],
    '2016': [20000, 50000, 70000, 100000, 125000],
    '2017': [50000, 60000, 70000, 80000, 90000]
}

sales = pd.DataFrame(sales_data, index=['Madhu', 'Kusum', 'Kinshuk', 'Ankit', 'Shruti'])

# a) Display the row labels of Sales.
print("Row labels of Sales:")
print(sales.index)

# b) Display the column labels of Sales.
print("\nColumn labels of Sales:")
print(sales.columns)

# c) Display the data types of each column of Sales.
print("\nData types of Sales columns:")
print(sales.dtypes)

# d) Display the dimensions, shape, size and values of Sales.
print("\nDimensions of Sales:")
print(sales.shape)

print("\nSize of Sales:")
print(sales.size)

print("\nValues of Sales:")
print(sales)

# e) Display the last two rows of Sales.
print("\nLast two rows of Sales:")
print(sales.iloc[-2:])

# f) Display the first two columns of Sales.
print("\nFirst two columns of Sales:")
print(sales.iloc[:, :2])

# g) Create a dictionary using the following data. Use this dictionary to create a DataFrame Sales2.
sales_data_2018 = {
    '2018': [160000, 110000, 50000, 340000, 900000]
}

sales2 = pd.DataFrame(sales_data_2018, index=['Madhu', 'Kusum', 'Kinshuk', 'Ankit', 'Shruti'])

# h) Check if Sales2 is empty or it contains data.
if 'sales2' in locals() and not sales2.empty:
    print("\nSales2 contains data")
else:
    print("\nSales2 is empty")

print("\nSales2:")
print(sales2)

# 2. Dataset

# Load the Iris dataset from seaborn
df = sns.load_dataset('iris')

# Display the first few rows of the dataset
print("\nFirst few rows of the dataset:")
print(df.head())

# 3. Summary Statistics and Visualizations
# Summary statistics of the dataset
print("\nSummary Statistics:")
print(df.describe())

# Checking for missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Pairplot to see relationships between features and species
sns.pairplot(df, hue='species')
plt.suptitle('Pairplot of Iris Dataset', y=1.02)
plt.show()

# Box plots to visualize the distribution of each feature by species
plt.figure(figsize=(12, 6))
sns.boxplot(x='species', y='sepal_length', data=df)
plt.title('Box Plot of Sepal Length by Species')
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(x='species', y='sepal_width', data=df)
plt.title('Box Plot of Sepal Width by Species')
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(x='species', y='petal_length', data=df)
plt.title('Box Plot of Petal Length by Species')
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(x='species', y='petal_width', data=df)
plt.title('Box Plot of Petal Width by Species')
plt.show()

# Violin plot for a detailed distribution view
plt.figure(figsize=(12, 6))
sns.violinplot(x='species', y='petal_length', data=df)
plt.title('Violin Plot of Petal Length by Species')
plt.show()

# Correlation matrix - FIX: exclude the 'species' column which is causing the error
plt.figure(figsize=(10, 6))
numeric_df = df.select_dtypes(include=[np.number])  # Select only numeric columns
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Iris Features')
plt.show()

# 4. Data Insights
# Checking for significant differences between species using groupby
grouped = df.groupby('species').mean()
print("\nMean Values by Species:")
print(grouped)

# Insights can be derived from the visualizations
# - Identify any patterns or significant differences between species in terms of features
# - Look for any outliers that are notable, especially in the box plots

# 5. Conclusion
# Based on the descriptive analytics performed:
# - Iris-setosa has the smallest petal lengths and widths, making it distinct from other species.
# - Iris-virginica tends to have the largest petal lengths and widths, separating it clearly from the other two species.
# - Petal length and width are strong differentiators between species.
# - Potential outliers are noted in the petal width of Iris-setosa.
```
